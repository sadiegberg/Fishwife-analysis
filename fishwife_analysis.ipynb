```python
plt.figure()
plt.plot(retail['year'], retail['store_count'], marker='o', color='green', linewidth=2)
plt.title('Fishwife Retail Store Expansion (2021–2024)', fontsize=14)
plt.ylabel('Number of Stores')
plt.xlabel('Year')
plt.tight_layout()
plt.savefig('visuals/retail_expansion.png', bbox_inches='tight', dpi=150)
plt.show()
print("Chart saved to visuals/")
```

```python
trends['Time'] = pd.to_datetime(trends['Time'])

plt.figure()
plt.plot(trends['Time'], trends['Fishwife Tinned Seafood Co'], 
         color='steelblue', linewidth=2, label='Fishwife Tinned Seafood Co')
plt.plot(trends['Time'], trends['tinned fish'], 
         color='coral', linewidth=2, label='tinned fish')
plt.plot(trends['Time'], trends['canned fish'], 
         color='gray', linewidth=2, label='canned fish')
plt.title('Google Search Interest Over Time (US)', fontsize=14)
plt.ylabel('Search Interest (0–100)')
plt.xlabel('Date')
plt.legend()
plt.tight_layout()
plt.savefig('visuals/google_trends.png', bbox_inches='tight', dpi=150)
plt.show()
print("Chart saved to visuals/")
```

```python
tiktok['date'] = pd.to_datetime(tiktok['date'])
tinnedfish = tiktok[tiktok['hashtag'] == '#tinnedfish']

plt.figure()
plt.bar(tinnedfish['date'].astype(str), tinnedfish['view_count_millions'], 
        color='orchid', width=0.4)
plt.title('#TinnedFish TikTok Views Over Time (Millions)', fontsize=14)
plt.ylabel('Views (Millions)')
plt.xlabel('Date')
plt.tight_layout()
plt.savefig('visuals/tiktok_views.png', bbox_inches='tight', dpi=150)
plt.show()
print("Chart saved to visuals/")
```

```python
collabs['date'] = pd.to_datetime(collabs['date'])

color_map = {
    'Food': 'steelblue',
    'Fashion/Lifestyle': 'orchid',
    'Food/Local': 'seagreen',
    'Art/Lifestyle': 'goldenrod',
    'Lifestyle/Beverage': 'coral'
}
collab_colors = collabs['partner_type'].map(color_map).fillna('gray')

plt.figure(figsize=(12, 6))
plt.scatter(collabs['date'], collabs['partner'], 
            c=collab_colors, s=200, zorder=5)
plt.title('Fishwife Brand Collaborations Over Time', fontsize=14)
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('visuals/collab_timeline.png', bbox_inches='tight', dpi=150)
plt.show()
print("Chart saved to visuals/")
```

```python
plt.figure()
plt.plot(market['year'], market['us_market_size_usd_billions'], 
         marker='o', color='darkorange', linewidth=2)
plt.title('U.S. Tinned Fish Market Size — Actual + Projected (USD Billions)', fontsize=14)
plt.ylabel('Market Size (USD Billions)')
plt.xlabel('Year')
plt.tight_layout()
plt.savefig('visuals/market_size.png', bbox_inches='tight', dpi=150)
plt.show()
print("Chart saved to visuals/")
```

```python
fig, ax1 = plt.subplots(figsize=(10, 5))

ax1.bar(revenue['year'], revenue['revenue_usd'], 
        color='steelblue', alpha=0.6, label='Revenue', width=0.4)
ax1.set_ylabel('Revenue (USD)', color='steelblue', fontsize=12)
ax1.yaxis.set_major_formatter(
    mticker.FuncFormatter(lambda x, _: f'${x/1e6:.1f}M')
)
ax1.tick_params(axis='y', labelcolor='steelblue')

ax2 = ax1.twinx()
ax2.plot(retail['year'], retail['store_count'], 
         color='coral', marker='o', linewidth=2, label='Store Count')
ax2.set_ylabel('Retail Store Count', color='coral', fontsize=12)
ax2.tick_params(axis='y', labelcolor='coral')

plt.title('Fishwife — Revenue vs. Retail Expansion', fontsize=14)
fig.tight_layout()
plt.savefig('visuals/revenue_vs_retail.png', bbox_inches='tight', dpi=150)
plt.show()
print("Chart saved to visuals/")
```

## Conclusion

The data supports the thesis that Fishwife didn't just sell fish — it sold an identity.

- **Revenue** grew 10x in 4 years ($750K → $7.6M) driven almost entirely by organic social media until 2023
- **Google Trends** shows "tinned fish" rising sharply alongside Fishwife's growth while "canned fish" stayed flat — proof that Fishwife helped rebrand an entire category
- **Collaborations** evolved from art/lifestyle partners (2021) → food brand collabs (2022) → fashion crossovers (2025), showing a deliberate shift from food brand to lifestyle brand
- **Retail expansion** from 0 to 1,891 stores mirrors revenue growth, confirming that social-driven brand awareness translated directly into retail demand
- **TikTok** with 91M+ views on #tinnedfish shows the cultural moment Fishwife both rode and created

Fishwife's growth story is a case study in how strong aesthetic branding, community building, and strategic collaborations can transform a commodity product into a cultural movement.
